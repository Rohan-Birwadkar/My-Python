```python
#Display function for the board
def display_board(board):
    
    a=[" "," "," "," "," "," "," "," "," "]
    for i in range(0,9):
        a[i]=board[i]
    
    #tab approx 8 spaces
    print("\t"+"|"+"\t"+"|")
    print(f"    {a[6]}   "+"|"+f"   {a[7]}   "+"|"+f"    {a[8]}    ")
    print("________|_______|_________")
    print("\t"+"|"+"\t"+"|")
    print(f"    {a[3]}   "+"|"+f"   {a[4]}   "+"|"+f"    {a[5]}    ")
    
    print("________|_______|_________")
    print("\t"+"|"+"\t"+"|")
    print(f"    {a[0]}   "+"|"+f"   {a[1]}   "+"|"+f"    {a[2]}    ")
    print("\t"+"|"+"\t"+"|")
```

```python
#player input and marker assignment
def player_input():
    p1=""
    p2=""
    print("\n"+"Hello PLayers,"+"\n"+
          "Welcome To Tic=Tac-Toe"+"\n"+
          "Player number 1 has to choose a character between X and O"+"\n"+
         "And the remaining character will be assigned to Player number 2"+"\n"+
          "Player number 1 are you ready??!")
    y=input("Y or N")
    if y.lower() in ['n','no']:
        print("Okey quitting")
        return True
    while True:
        clear_output()
        p1=input("Player 1, Please choose Either X or O ")
        if p1.lower()=='x':
            print("Then its decided that"+"\n"+"Player number 1 is: X"+"\n"+"Player number 2 is: O")
            p1='X'
            p2="O"
            return [p1,p2]
            break
        elif p1.lower()=="o":
            print("Then its decided that"+"\n"+"Player number 1 is: O"+"\n"+"Player number 2 is: X ")
            p1="O"
            p2="X"
            return [p1,p2]
            break
        else:
            print("Player number 1, please choose only among the two options")
            continue
                
```

```python
#assignment of marker to position on board
def place_marker(place_marker_board, place_marker_marker, place_marker_position):
    place_marker_board.pop(place_marker_position)
    place_marker_board.insert(place_marker_position,place_marker_marker)
```

```python
#free space on board check function
def space_check(space_check_board, space_check_position):
    if space_check_board[space_check_position]==" ":
        return True
    else:
        return False
```

```python
#full board checker
def full_board_check(full_board_check_board):
    is_empty=9
    is_full=0
    for i in range(0,9):
        if full_board_check_board[i]!=" ":
            is_full+=1
    if is_empty>is_full:
        return False
    return True
```

```python
#mark placer function 
#check if free
#check if full
#check if win
def player_choice(player_choice_board):
    player_choice_pos=0
    while True:
        print("Enter the position you desire to mark")
        player_choice_pos=input("here"+"\n")
        #digit check
        if player_choice_pos.isdigit():
            player_choice_pos=int(player_choice_pos)
            player_choice_pos=player_choice_pos-1
            #range check
            if player_choice_pos in range(0,9):
                #space check
                if space_check(player_choice_board,player_choice_pos):
                    return player_choice_pos
                else:
                    if full_board_check(player_choice_board):
                        print("board full lets start again")
                        return 10
                    print("Position alredy used so choose again")
                    continue
            else:
                print("Position out of range")
                continue
        else:
            print("Enter number only")
            continue
    
```
